Q2: In the above question, now scrape the following details of each product listed in first 3 pages of your search results and 
    save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products
    available under that product name. Details to be scraped are: "Brand Name", "Name of the Product", "Price",
    "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”. In case, if any of the details are missing for any
    of the product then replace it by “-“.

```python
#creating the empty list
brand=[]
description=[]
price=[]
Expected_delivery=[]


time.sleep(3)

#scrapping the required details
start=0
end=3
for page in range(start,end):#for loop for scrapping 4 page
    brands=driver.find_elements_by_xpath('//span[@class="a-size-medium a-color-base a-text-normal"]')#scraping brands name by class name='_2WkVRV'
    for i in brands:
        brand.append(i.text)#appending the text in Brand list
    desc=driver.find_elements_by_xpath('//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')#scraping description from the xpath
    for i in desc:
        description.append(i.text)#appending the description in list
    prices=driver.find_elements_by_xpath("//span[@class='a-price-whole']")# scraping the price from the xpath
    for i in prices:
        price.append(i.text)
    Expected_delivery_date=driver.find_elements_by_xpath("//span[@class='a-color-base a-text-bold']")# scraping the price from the xpath
    for i in Expected_delivery_date:
        Expected_delivery.append(i.text)
        
        
    nxt_button=driver.find_elements_by_xpath("//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']")#scraping the list of buttons from the page
    try:
        driver.get(nxt_button[1].get_attribute('href'))#getting the link from the list for next page
    except:
        driver.get(nxt_button[0].get_attribute('href'))
```

```python
len(brand), len(description), len(price), len(Expected_delivery)
```

```python
guiter_list = pd.DataFrame()
guiter_list['Brand'] = brand[0:59]
guiter_list['Product Description'] = description[0:59]
guiter_list['Price'] = price[0:59]
guiter_list['Expected_Delivery'] = Expected_delivery
guiter_list
```

```python
guiter_list.to_csv('C:/Users/humay/OneDrive/Desktop/Data Science/Flip Robo Internship/guiter_list.csv')
```
